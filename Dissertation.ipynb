```python
#data clean part
import pandas as pd
import re
#è¯»å–æ–‡ä»¶
#df = pd.read_csv('/Users/mmy/Desktop/TestData.csv')
file_path=r'/Users/mmy/Desktop/merge_data.csv'
#Set encoding format
df=pd.read_csv(file_path,encoding='utf-8')
#data clean
def data_clean(df):
    df=df.drop(columns=['id','bid',])
    
    df.columns=['text','headline_url','original_image_url','video_url','location','date','tool','likes','comment_number','forwarding_number','topic','_user']
    #Clean up error messages
    df=df.drop(columns=['location'])
    #columnä¿¡æ¯åˆ¤æ–­ï¼Œæœ‰åˆ™ä¸º1ï¼Œæ²¡æœ‰ä¸º0
    for i in range(len(df)):
        if str(df['_user'][i])=='nan':
            df['_user'][i]=0
        else:
            df['_user'][i]=1
        if str(df['headline_url'][i])=='nan':
            df['headline_url'][i]=0
        else: 
            df['headline_url'][i]=1
            
        if str(df['original_image_url'][i])=='nan':
            df['original_image_url'][i]=0
        else: 
            df['original_image_url'][i]=1
            
        if str(df['video_url'][i])=='nan':
            df['video_url'][i]=0
        else:
            df['video_url'][i]=1
            
        if str(df['tool'][i])=='nan':
            df['tool'][i]=0
        else:
            df['tool'][i]=1
            
        if str(df['topic'][i])=='nan':
            df['topic'][i]=0
        else:
            df['topic'][i]=1
    #
    df['date'] = pd.to_datetime(df['date'],format='%Y-%m-%d %H:%M:%S')#å°†è¯»å–çš„æ—¥æœŸè½¬ä¸ºdatatimeæ ¼å¼
    #x=[i.day for i in df["date"]]#éå†æ—¶é—´å˜é‡ï¼Œæå–å¹´ä»½ï¼Œä½¿ç”¨datatimeæ¨¡å—çš„å±æ€§å’Œæ–¹æ³•
    df['day']=None
    df['month']=None
    for i in range(len(df["date"])):
        #x=int(df["date"][i].day)
        df['day'][i]=df["date"][i].day
        df['month'][i]=df["date"][i].month
    
    print(df.day)
    print(df.month)
    df=df.drop(columns=['date'])
    
    #Article content Keyword extraction
    df['word_number']=None
    df['sentence_number']=None
    df['emoji_number']=None
    df['word_per_sentence']=None
    df['digit_number']=None
    pattern_word_number = re.compile(r'[0-9A-Za-z\u4e00-\u9fa5]')
    pattern_sentence_number = re.compile(r'[\uff01\u3002\uff1b\uff0c\uff1a\u2026\uff1f]')
    pattern_emoji_number = re.compile(r'[^0-9A-Za-z\u4e00-\u9fa5\u007e\u2053\u3002\uff1b\uff0c\uff1a\u201c\u201d\uff08\uff09\u3001\uff1f\u300a\u300b\u2026\u0020-\u002f\u003a-\u0040]')
    pattern_digit_number = re.compile(r'[0-9]')
    for i in range(len(df['text'])):
        result_a = pattern_word_number.findall(df['text'][i])
        df['word_number'][i]=len(result_a)#Word count
        
        result_b = pattern_sentence_number.findall(df['text'][i])
        df['sentence_number'][i]=len(result_b)#sentence count
        
        if df['sentence_number'][i]<1: #Solve zero errors
            df['sentence_number'][i]=1
            
        result_c = pattern_emoji_number.findall(df['text'][i])
        df['emoji_number'][i]=len(result_c)#emoji count
        
        df['word_per_sentence'][i]=df['word_number'][i]/df['sentence_number'][i]
        
        result_d = pattern_digit_number.findall(df['text'][i])
        df['digit_number'][i]=len(result_d)#digit count
    
    return df
    
#ä»¥ä¸Šdata processå®Œæˆ
#æ–°çš„è®­ç»ƒæ¨¡å‹

####jieba åˆ†æï¼Œå»ä¸­æ–‡åœç”¨è¯
df=data_clean(df)
print(df['digit_number'])
print(df.dtypes)
df.to_csv('/Users/mmy/Desktop/NEWWWWWW.csv',encoding='utf_8_sig')
print('finished')
```
